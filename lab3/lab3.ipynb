```python
from scipy.integrate import odeint
g = lambda y, t: 2 - y**2 - y

fig = plt.gcf()
fig.set_size_inches(14, 12)


X = np.linspace(1, 7, 32)
Y = np.linspace(0, 5, 32)

U = [1 for i in range(32)]
V = [f(i) for i in Y]

XX, YY = np.meshgrid(X, Y)


UU, VV = np.meshgrid(U, V)


plt.quiver(XX, YY, UU, VV, units='xy', width=0.004, scale=20)


for y0 in np.linspace(0, 5, 10):
    t = np.linspace(1, 7)
    r = odeint(g, y0, t)
    plt.plot(t, r)

plt.show()
```

### Task 2.3
Find all inflection points of integral curves

Inflection points of a function is defined by zeroes of its second derivative

$$y' = f(x, y)$$
$$y'' = \frac{\partial f}{\partial x}dx + \frac {\partial f}{\partial y}dy = 
f^{'}_{x} + f^{'}_{y}y'$$
Substituting $y' = f(x, y)$

$$f^{'}_{x} + f^{'}_{y}f(x, y) = f^{'}_{x} + f^{'}_{y}f = 0$$
$$f_{x}^{'} + f_{y}^{'}f = 0$$
$$-(1 + 2y)(2 - y - y^2) = 0$$
$$y = -\frac{1}{2}, y = -2, y = 1$$ 

Hence $y_i, i=1,..,3$ are geometric place of possible inflection points 

```python
fig = plt.gcf()
fig.set_size_inches(10, 8)

for y0 in np.linspace(-1.995, 5, 30):
    for t0 in np.linspace(1, 2, 5):
        t = np.linspace(t0, t0 + 5)
        r = odeint(g, y0, t)
        plt.plot(t, r)
plt.grid(axis='both')
t = np.linspace(1, 7, 2)
plt.plot(t, (-0.5, -0.5), 'r--', t, (-2, -2), 'r--', t, (1, 1), 'r--', linewidth=0.8)
plt.show()
```

As One can see, line $y \equiv -\frac{1}{2}$ is geometric place of inflection points

## Task 2.4

**Let $y(x, y_0), x \in I_{y_0}$ be a non-continuable solution, $y(0, y_0) = y_0$, find interval $I_{y_0}$**

$$y' = 2 - y^2 - y $$
$$\frac{dy}{(y+2)(y-1)} = dx$$
$$\int{\frac{dy}{(y+2)(y-1)}} = \int{dx}$$
$$\int{\frac{dy}{(y+2)(y-1)}} = \frac{1}{3}\int{\bigg( \frac{1}{y - 1} - \frac{1}{y + 2} \bigg) dy}$$
$$\ln{\bigg|\frac{y - 1}{y + 2}\bigg|} = 3x + C$$
$$\frac{y - 1}{y + 2} = C_{1}e^{3x}$$
(Considering $y(0, y_0) = y_0$)
$$\frac{y_0 - 1}{y_0 + 2} = C$$
Now let's derive $y(x, C)$
$$\frac{y - 1}{y + 2} = C_{1}e^{3x}$$
$$y - 1 = yCe^{3x} + 2Ce^{3x}$$
$$y(1 - Ce^{3x}) = 2Ce^{3x} + 1$$
$$y = \frac{2Ce^{3x} + 1}{1 - Ce^3x} = - \frac{2Ce^{3x} - 2 + 3}{Ce^{3x} - 1} = - \bigg(2 + \frac{3}{Ce^{3x} - 1} \bigg) = - 2 - \frac{3}{Ce^{3x} - 1} = - 2 + \frac{3}{C_1e^{3x} + 1} $$(minus sign in denominator absorbed)

Substituting $C = \frac{y_0 - 1}{y_0 + 2}$ get
$$y(x, y_0) = - 2 - \frac{3}{\frac{y_0 - 1}{y_0 + 2} e^{3x} + 1}$$ and
$$y \equiv -2$$

if $y \equiv -2$, then $I_{y_0} = \mathbb{R}$, if $y \ne -2$ problem may occur when denominator turns into zero.
$$\frac{y_0 - 1}{y_0 + 2} e^{3x} + 1 = 0, e^{3x} = -\frac{y_0 + 2}{y_0 - 1}$$, if $y_0 = 1 \implies x \in \emptyset$ else:

$e^{3x} \gt 0 \implies \frac{y_0 + 2}{y_0 - 1} \lt 0 \iff y \in (-2, 1) \implies$ if $y_0 \in (-2, 1)$ then $I_{y_0} = \mathbb{R}$

- if $y_0 < -2:$ then $x_0 = \frac{1}{3} \ln{\frac{y_0 - 2}{y_0 + 1}} = \frac{1}{3} \ln{\big(1 + \frac{3}{y_0 + 1}\big)}$; $y_0 < -2 \implies y_0  < - 1 \implies \frac{3}{y_0 + 1} < 0 \implies \frac{1}{3} \ln{\big(1 + \frac{3}{y_0 + 1}\big)} < 0 \implies I_{y_0} = (x_0, +\infty)$

- if $y_0 > 1$ then $x_0 = \frac{1}{3} \ln{\big(1 + \frac{3}{y_0 + 1}\big)}$; $y_0 > 1 \implies y_0 > 1 \implies \frac{3}{y_0 + 1} > 0 \implies \frac{1}{3} \ln{\big(1 + \frac{3}{y_0 + 1}\big)} > 0 \implies I_{y_0} = (-\infty, x_0)$

## Task 3.1
For given differential equation $y' = y^2 + x - 2$ plot it's vector field

```python
def plotVectorField(f, xMin, xMax, yMin, yMax):
    fig = plt.gcf()
    fig.set_size_inches(12, 10)

    X = np.linspace(xMin, xMax, 64)
    Y = np.linspace(yMin, yMax, 64)
    
    XX, YY = np.meshgrid(X, Y)
    
    UU = np.cos(np.arctan(f(XX, YY)))
    VV = np.sin(np.arctan(f(XX, YY)))
    C = [[ f(x, y)*64 for x in X] for y in Y]
    plt.quiver(XX, YY, UU, VV, C, units='xy', width=0.02, scale=5)
    plt.title("Warmer color - higher speed")
    
f = lambda x, y: y**2 + x - 2    
plotVectorField(f, -3, 3, -6, 6)
plt.show()
```

```python
plotVectorField(f, -3, 5, -6, 6)

g = lambda x: (2 + k - x)**0.5

vf = np.vectorize(g)

for k in range(1, 4):
    x = np.linspace(-3, 2 + k)
    y = vf(x)
    leg = "k=" + str(k)
    plt.plot(x, y, 'r--', label=leg)
    plt.plot(x, -y, 'r--')
plt.show()
```

## Task 3.6
Find and draw geometric place of integral curves family inflection points

$$y'' = f_{x}^{'} + f_{y}^{'}f = 0$$
$$1+2y(y^2 + x - 2) = 0$$
$$2y^3 + 2xy - 4y + 1 = 0$$

Last equation defines geometric place of possible inflection points

Here we will use sympy to plot implicit graph, after it we will gather points from it and plot it upon our vector field

```python
from sympy import var, plot_implicit

res = []
var('x y')
s = plot_implicit(2*y**3 + 2*x*y - 4*y + 1).__dict__["_series"]
intervals = s[0].get_raster()[0]
for v in intervals:
    x, y = v[0].start, v[1].start
    res.append((x, y))
```

```python
plotVectorField(f, -5, 5, -6, 6)

fig = plt.gcf()
fig.set_size_inches(12, 10)

X, Y = zip(*res)
line = plt.scatter(X, Y, c='r', s=0.005)
plt.legend((line, ), ("Integral curves inflection points",))
plt.show()
```
